```python
import pandas as pd
import numpy as np
```

```python
countries = pd.read_csv(r"C:\Users\KondasaniBalaji\ai_training\Week1\Day5\Countries.csv")
countries.head()
```

```python
# country with  highest population
countries[countries["population"]==countries["population"].max()][["country","population"]]
```

```python
# capital of country with highest population
countries[countries["population"]==countries["population"].max()][["country","capital_city","population"]]
```

```python
# country with least population
countries[countries["population"]==countries["population"].min()][["country","population"]]
```

```python
countries[countries["population"]==countries["population"].min()][["country","capital_city","population"]]
```

```python
# top d countries with highest democratic score
countries.sort_values(by="democracy_score",ascending=False).head(4)[["country","capital_city","democracy_score"]]
```

```python
# get total regions
countries["region"].value_counts().count()
```

```python
# countries in eastern europe region
countries[ countries["region"] == "Eastern Europe"][["country","population"]].sort_values(by="population",ascending=False)
```

```python
#political leader of second highest populated country
countries.nlargest(n=2,columns="population")[["political_leader","country"]].iloc[1]
```

```python
# countries whose political leaders are unknown
countries[countries["political_leader"].isna()][["country","political_leader"]]
```

```python
# how many countries have republic in their full name
countries[countries["country"].str.contains("Republic")][["country","political_leader"]]
```

```python
countries[ countries["region"].str.contains("Africa") ][["country","region","population"] ].nlargest(n=2,columns="population").iloc[1]
```
