```python
import pandas as pd
import numpy as np
```

```python
#loading a dataset -- Anime CSV

anime = pd.read_csv(r"C:\Users\KondasaniBalaji\ai_training\Week1\Day5\anime.csv")

anime.head()
```

```python
# a new column for feature extraction

def extract_episodes(title):
    data=""
    i=0
    while i< len(title):
        if title[i]=="(":
            i+=1
            while i<len(title) and title[i]!=")":
                if title[i].isnumeric():
                    data+=title[i]
                i+=1
            break
        i+=1
    return int(data)
                        
```

```python
anime["episode_count"]=anime["Title"].apply(extract_episodes)
anime
```

```python
def extract_timestamp(title):
    i=0
    data=""
    while i < len(title):
        if title[i]==")":
            data+=title[i+1:i+20]
            break
        i+=1
    return int(data)
```

```python
anime["Release"] = anime["Title"].apply(extract_timestamp)
anime.head()
```

```python
# convert timestamp to months
anime[['start', 'end']] = anime['Release'].str.split(' - ', expand=True)

anime['start'] = pd.to_datetime(anime['start'], format='%b %Y')
anime['end'] = pd.to_datetime(anime['end'], format='%b %Y')

anime['duration_months'] = (
    (anime['end'].dt.year - anime['start'].dt.year) * 12
    + (anime['end'].dt.month - anime['start'].dt.month)
)

anime[['duration_months']].head()
```
