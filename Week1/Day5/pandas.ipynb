FINDING MISSING DATA

```python
data = {
    "A": [1, 2, np.nan, 4, 5],
    "B": [6, 2, 3, 4, 5],
    "C": [1, 2, 3, np.nan, np.nan],
    "D": [2, np.nan, np.nan, np.nan, 5]
}

df3 = pd.DataFrame(data)
df3
```

```python
# finding missing data -- isna() -- column wise check
print(df3.isna().sum(),end="\n\n")

# checking if null values exist
print(df3.isna().any())
```

```python
# removing missing data -- dropna() -- row wise removal , for inplace modification use inplace = True

df3.dropna()

# thresh parameter -- removes rows that dont have thresh no of non null values
df3.dropna(thresh=3)
```

```python
#filling missing values -- fillna()

df3.fillna(0)

# filling column with specific values

values = {"A":0,"B":3,"C":9,"D":10}
df3.fillna(value=values)

# filling column values with mean

df3.fillna(df3.mean())
```

MERGING , JOINING AND CONCATENATION


```python
employees = {
    "employee_id": [1, 2, 3, 4, 5],
    "name": ["John", "Anna", "Peter", "Linda", "Bob"],
    "department": ["HR", "IT", "Finance", "IT", "HR"]
}
salaries = {
    "employee_id": [1, 2, 3, 6, 7],
    "salary": [60000, 80000, 65000, 70000, 90000],
    "bonus": [5000, 10000, 7000, 8000, 12000]
}
df4 = pd.DataFrame(employees)
df6 = pd.DataFrame(salaries)

print(df4,"\n\n",df6)
```

```python
# merging two dataframes
pd.merge(df4,df6,on="employee_id",how="outer")
```

```python
f7 = {
    "A": ["A0", "A1", "A2"],
    "B": ["B0", "B1", "B2"],
    "C": ["C0", "C1", "C2"]
}

f8 = {
    "A": ["A3", "A4", "A5"],
    "B": ["B3", "B4", "B5"],
    "C": ["C3", "C4", "C5"]
}

df7 = pd.DataFrame(f7)
df8 = pd.DataFrame(f8)

print(df7,"\n\n",df8)
```

```python
#concatenating two dataframes -- 1 arg -- list of data values to e merged

pd.concat([df7,df8],axis=1)
```

```python
f9 = {
    "name": ["Alice", "Bob", "Charlie"]
}
index1 = [1, 2, 3]

f10 = {
    "score": [85, 90, 75]
}
index2 = [2, 3, 4]

df9 = pd.DataFrame(f9,index1)
df10 = pd.DataFrame(f10,index2)
```

GROUP BY AND AGGREGATION

```python
data = {
    "Category": ["A", "B", "A", "B", "A", "B", "A", "B"],
    "Store": ["S1", "S1", "S2", "S2", "S1", "S2", "S2", "S1"],
    "Sales": [100, 200, 150, 250, 120, 180, 200, 300],
    "Quantity": [10, 15, 12, 18, 8, 20, 15, 25],
    "Date": pd.date_range("2023-01-01", periods=8)
}

df11=pd.DataFrame(data)
print(df11)
```

```python
# group by category and calculate sum of sales

print(df11.groupby("Category")["Sales"].sum())

# group by category and calculate sum of sales

df11.groupby("Store")["Sales"].sum()
```

```python
# group by multiple columns and perform aggregation

gd=df11.groupby(["Category","Store"])["Sales"].mean()
df11["Sales"].mean()
```

```python
# .agg() combines multiple statistical operations
df11["Sales"].agg(["mean","median","mean","min","max","count","std"])
```

```python
#pivot tables
data = {
    "Date": pd.date_range("2023-01-01", periods=20),
    "Product": ["A", "B", "C", "D"] * 5,
    "Region": ["North", "South", "East", "West", "North", "South", "East", "West", "North", "South", "East", "West", "North", "South", "East", "West", "North", "South", "East", "West"],
    "Sales": np.random.randint(100, 1000, 20),
    "Units": np.random.randint(10, 100, 20),
    "Rep": ["John", "Mary", "Bob", "Alice", "John", "Mary", "Bob", "Alice", "John", "Mary", "Bob", "Alice", "John", "Mary", "Bob", "Alice", "John", "Mary", "Bob", "Alice"]
}

df = pd.DataFrame(data)

df["Month"] = df["Date"].dt.month_name()
df["Quarter"] = "Q" + df["Date"].dt.quarter.astype(str)
df
```

```python
#values -- column in which aggregation is being applied
#index -- rows labelling or naming
#columns -- columns labelling or naming

pd.pivot_table(df,values="Sales",index="Region",columns="Product",aggfunc="median")

#pivot table on multiple columns

pd.pivot_table(df,values=["Sales","Units"],index="Region",columns="Product")
```

CROSSTABS
