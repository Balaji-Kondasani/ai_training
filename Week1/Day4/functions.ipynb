```python
# **args and **kwargs

def variable_args(*args):
    '''
    Allow us to pass variable no of non-keyword arguments to a function
    '''
    product=1
    for i in args:
        product*=i
    print(args)
    return product
```

```python
variable_args(2,4,9,8,4,9,3,2,4)
```

```python
# kwargs

def display(**kwargs):
    '''
    kwargs -- allows us to pass a variable no of non_keyword args as key value pairs of python dict
    '''
    print(display.__doc__)
    for (key,value) in kwargs.items():
        print(f"{key} -- {value}")
```

```python
display(a=1,b=2,c=3,d=4,e=9,z=10)
```

FUNCTIONS ARE FIRST CLASS CITIZENS

What First-Class Means in Practice?

    An entity is considered first-class if it can perform the following four operations: 

    1. Assign to a Variable: You can store a function in a variable without calling it immediately.
    
    2. Pass as an Argument: You can send a function into another function (this is how "higher-order functions" like map or filter work).

    3. Return from a Function: A function can generate and return another function as its result.

    4. Store in Data Structures: You can place functions inside lists, dictionaries, or sets

LAMBDA FUNCTIONS

```python
# Lambda Functions -- a single line anonymous function expression used mostly with higher order functions

# can take any no of arguments , has only one expression , not reusable

# used with higher order functions

sum=lambda  a,b,c : a+b+c

print(sum(2,3,6),"\n")

#check if a string has "a"

check_char = lambda str : True if "a" in str else False

print(check_char("India"))

#check if no is odd or even

check_even_odd = lambda num : "Even" if num%2==0 else "Odd"

check_even_odd(6)
```

HIGHER ORDER FUNCTIONS


```python
# Function that receives another function as argument

def square(x):
    return x**2

def modify(f,l):
    result=[]
    for i in l:
        result.append(f(i))
    print(result)

L=[1,2,3,4,5,6]
modify(square,L)
```

MAP FUNCTION

```python
# map function in python
list1=[1,2,3,6,7,8]

squared_nums = list(map(lambda num:num*num,list1))
print(squared_nums)
```

```python
# odd/even labelling in python

list1=[1,2,3,4,5,6]

result=list(map(lambda num: "Even" if num%2==0 else "Odd",list1))

print(result)
```

```python
# fetch names from a list of dictionaries

users=[
    
    {
        "name":"balaji kondasani",
        "age":21,
        "email":"b@gmail.com"
    },
    {
        "name":"Moonlander",
        "age":74,
        "email":"m@gmail.com"
    },
    {
        "name":"Milky Way Galaxy",
        "age":16,
        "email":"mwg@gmail.com"
    }
]

result=list(map(lambda users:users["name"],users))
print(result)
```

FILTER

```python
# filter numbers greater>5
list1=[3,2,7,9,11,2,3]

result = list(filter(lambda x:x>5,list1))

print(result,"\n\n")

# filter strings starting with a
list1 = ["apple","america","amsterdam","India","AssertionError"]

result = list(filter(lambda str : str.lower().startswith("a"),list1))

print(result)
```

REDUCE

```python
# CALCULATE SUM OF A LIST

import functools
list1 = [1,3,6,8,10]
result=functools.reduce(lambda x,y:x+y,list1)
print(result)

# Find maximum number in a list

list2 = [i for i in range(6)]
result = functools.reduce(lambda x,y: x if x>y else y,list2)
print(result)

```
