Broadcasting

Broadcasting talks about how numpy treats or handles arrays of different shapes during arithmetic operations

Smaller array is broadcasted across larger array so that they have compatible shapes

In smaller array duplication of existing rows happen horizontally and vertically


BroadCasting Rules:
    1. Make the two arrays have the same number of dimensions.
    2. If the numbers of dimensions of the two arrays are different, add new dimensions with size 1 to the head of the array with the smaller dimension.

    3. Make each dimension of the two arrays the same size.
    
    4. If the sizes of each dimension of the two arrays do not match, dimensions with size 1 are stretched to the size of the other array.
    
    5. If there is a dimension whose size is not 1 in either of the two arrays, it cannot be broadcasted, and an error is raised.

```python
a1 = np.arange(12).reshape(3,4)
a2 = np.arange(4).reshape(1,4)

print(a1+a2)

a3 = np.arange(6).reshape(3,2)
a4 = np.arange(9).reshape(3,3)

try:
    print(a3+a4,"\n\n")
except Exception as e:
    print(f"a3 + a4 could not be performed as {e} Occured")
```

WORKING WITH MATHEMATICAL FORMULAS

```python
#SIGMOID -- formula -> 1 / 1+ e^-x

def sigmoid(array):
    return (1 / 1+ np.exp(-(array)))
sig=sigmoid(np.arange(10))
print(sig)
```

```python
#MEAN SQUARE ERROR

def msr(actual,predicted):
    return np.mean((predicted-actual)**2)

actual = np.random.randint(0,50,25)
predicted = np.random.randint(0,50,25)
result=msr(actual,predicted)

print(result)

```

SET FUNCTIONS
 1. np.union1d()
 2. np.intersect1d()
 3. np.setdiff1d()
 4. np.setxor1d()

```python
m = np.array([1,2,3,4,5])
n = np.array([3,4,5,6,7])

print(np.union1d(m,n))
print(np.intersect1d(m,n))
print(np.setdiff1d(m,n))
print(np.setdiff1d(n,m))
print(np.setxor1d(m,n))
```
